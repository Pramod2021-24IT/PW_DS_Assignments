### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a Python library used for creating a wide range of static, animated, and interactive plots. It is especially useful for data visualization, helping to create clear, informative graphics. 

### Five Plots in Matplotlib's Pyplot Module

1. **Line Plot**: Shows trends over time or continuous data.
   ```python
   plt.plot(x, y)
   ```

2. **Bar Plot**: Compares quantities across different categories.
   ```python
   plt.bar(categories, values)
   ```

3. **Histogram**: Displays the distribution of data.
   ```python
   plt.hist(data, bins)
   ```

4. **Scatter Plot**: Visualizes the relationship between two variables.
   ```python
   plt.scatter(x, y)
   ```

5. **Pie Chart**: Represents proportions of a whole.
   ```python
   plt.pie(sizes, labels=labels, autopct='%1.1f%%')
   ```

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np <br>
np.random.seed(3) <br>
x = 3 + np.random.normal(0, 2, 50) <br>
y = 3 + np.random.normal(0, 2, len(x))

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create scatter plot
plt.scatter(x, y)

# Add title and labels
plt.title('Scatter Plot of x vs y')
plt.xlabel('x values')
plt.ylabel('y values')

# Show plot
plt.show()
````

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data: <br>
import numpy as np <br>
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500]) <br>
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70]) <br>
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60]) <br>
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

````python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure and a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot each line in a subplot
axs[0, 0].plot(x1, y1, 'r-')  # Red line for plot 1
axs[0, 0].set_title('Line 1')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')

axs[0, 1].plot(x2, y2, 'g-')  # Green line for plot 2
axs[0, 1].set_title('Line 2')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')

axs[1, 0].plot(x3, y3, 'b-')  # Blue line for plot 3
axs[1, 0].set_title('Line 3')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')

axs[1, 1].plot(x4, y4, 'm-')  # Magenta line for plot 4
axs[1, 1].set_title('Line 4')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
````

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np <br>
company = np.array(["Apple", "Microsoft", "Google", "AMD"]) <br>
profit = np.array([3000, 8000, 1000, 10000])

````python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a figure and axis objects
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Vertical Bar Plot
axs[0].bar(company, profit, color='skyblue')
axs[0].set_title('Vertical Bar Plot')
axs[0].set_xlabel('Company')
axs[0].set_ylabel('Profit')

# Horizontal Bar Plot
axs[1].barh(company, profit, color='salmon')
axs[1].set_title('Horizontal Bar Plot')
axs[1].set_xlabel('Profit')
axs[1].set_ylabel('Company')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
````

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200) <br>
box2 = np.random.normal(90, 20, 200)

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a figure and axis object
plt.figure(figsize=(8, 6))

# Plot the box plot
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])

# Add title and labels
plt.title('Box Plot of Box 1 and Box 2')
plt.xlabel('Box')
plt.ylabel('Values')

# Show the plot
plt.show()
````
