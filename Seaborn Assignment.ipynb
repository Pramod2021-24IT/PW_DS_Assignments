### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
# 1. Histogram
# Function: seaborn.histplot()
# Use: Displays the distribution of a single variable. It helps to understand the frequency distribution of data points and visualize the shape of the distribution.

import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('iris')
sns.histplot(data['sepal_length'], bins=20, kde=True)
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
# 2. Box Plot
# Function: seaborn.boxplot()
# Use: Provides a visual summary of the data distribution, including the median, quartiles, and potential outliers. It is useful for comparing distributions between categories.

import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('iris')
sns.boxplot(x='species', y='sepal_length', data=data)
plt.title('Box Plot of Sepal Length by Species')
plt.show()
```

```python
# 3. Scatter Plot
# Function: seaborn.scatterplot()
# Use: Shows the relationship between two numerical variables. It helps to identify patterns, trends, and correlations between variables.

import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('iris')
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=data)
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.show()
```

```python
# 4. Pair Plot
# Function: seaborn.pairplot()
# Use: Creates a grid of scatter plots for all pairs of numerical variables in a dataset, along with histograms on the diagonal. It is useful for exploring relationships between multiple variables and detecting patterns.

import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('iris')
sns.pairplot(data, hue='species')
plt.title('Pair Plot of Iris Dataset')
plt.show()
```

```python
# 5. Heatmap
# Function: seaborn.heatmap()
# Use: Visualizes matrix-like data with color-coded cells. It is often used to display correlations or the frequency of events in a grid format, making it useful for heatmap-style data visualization.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = sns.load_dataset('flights')

# Pivot the dataset to get it in the correct format for a heatmap
pivot_data = data.pivot(index='month', columns='year', values='passengers')

# Create the heatmap
sns.heatmap(pivot_data, annot=True, fmt='d', cmap='YlGnBu')

# Add title and display the plot
plt.title('Heatmap of Flight Passengers')
plt.show()

```

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
df = sns.load_dataset('fmri')

# Create the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='timepoint', y='signal', hue='event', style='region')

# Add titles and labels
plt.title('Signal over Timepoint for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')

# Show the plot
plt.legend(title='Event and Region')
plt.show()
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
df = sns.load_dataset('titanic')

# Create a figure with two subplots
plt.figure(figsize=(14, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='pclass', y='age')
plt.title('Box Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='pclass', y='fare')
plt.title('Box Plot of Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
df = sns.load_dataset('diamonds')

# Plot a histogram for 'price' with hue based on 'cut'
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='price', hue='cut', multiple='stack', palette='viridis', bins=30)

# Set plot title and labels
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
df = sns.load_dataset('iris')

# Plot a pair plot
sns.pairplot(df, hue='species', palette='Set1')

# Show the plot
plt.show()
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
df = sns.load_dataset('flights')

# Pivot the dataset to get a matrix format, specify observed=False to avoid FutureWarning
pivot_table = df.pivot_table(index='month', columns='year', values='passengers', observed=False)

# Plot the heatmap
sns.heatmap(pivot_table, annot=True, fmt='g', cmap='YlGnBu')

# Add a title
plt.title('Heatmap of Flight Passengers')

# Show the plot
plt.show()
```
